# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
input_labels = ['red','black','red','green','black','yellow','white']
```

```python
encoder = LabelEncoder()
encoder.fit(input_labels)
```

```python
values = encoder.fit_transform(input_labels)
print("\nLabels =", input_labels)
print("Encoded values =", list(values))
```

```python
test_labels = ['green','red','black']
```

```python
encoded_values = encoder.transform(test_labels)
print("\nLabels =", test_labels)
print("Encoded values =", list(encoded_values))  
```

```python
encoded_values = [3,0,4,1]
```

```python
decoded_list = encoder.inverse_transform(encoded_values)
```

# Feature Importance with ExtratreeClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier(random_state=42)
```

```python
#all unnecessary features 
model.fit(X,y)
print(model.feature_importances_)
```

```python
model.score(X,y)
```

```python
fea_imp = pd.DataFrame({'imp':model.feature_importances_, 'col': X.columns})
fea_imp = fea_imp.sort_values(['imp', 'col'], ascending=[True, False]).iloc[-30:]
_ = fea_imp.plot(kind='barh', x='col', y='imp', figsize=(20, 10))
#plt.savefig('catboost_feature_importance4.png')
```
