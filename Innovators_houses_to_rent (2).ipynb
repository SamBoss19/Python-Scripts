### Data Aggregation With Group By.

```python
df1.columns
```

#### Highest Rent Per City on Average

```python
df4=df1.groupby('city')[['rent amount (R$)']].mean().sort_values('rent amount (R$)') #average rent per city,
df4
```

```python
df4.plot.bar(title='Average rent per city');
```

The Chart Shows that on the Average, Sao Paulo has the highest rent Amount.

#### Do Furnished apartments affect the Rent Prices ?

```python
df5=df.groupby(['city','furniture'])[['rent amount (R$)']].mean()
df5=df5.reset_index()
df5
```

```python
sb.barplot(x='city', y='rent amount (R$)', hue='furniture',data=df5);
```

Furnished Apartments have higher rents.

#### Do Pet Allowance in apartments affect the Rent ?

```python
df6=df.groupby(['city','animal'])[['total (R$)']].mean()
df6=df6.reset_index()
df6
```

```python
sb.barplot(x='city', y='total (R$)', hue='animal',data=df6);
```

Rents of Houses with Pet allowance are higher for all cities apart from Belo Horizonte.
